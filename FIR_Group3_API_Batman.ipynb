```python
import tweepy
import configparser
import pandas as pd

```

```python
# read configs
config = configparser.ConfigParser()
config.read('config_suha.ini')

api_key = config['twitter']['api_key']
api_key_secret = config['twitter']['api_key_secret']

access_token = config['twitter']['access_token']
access_token_secret = config['twitter']['access_token_secret']
```

```python
# authentication
auth = tweepy.OAuthHandler(api_key, api_key_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth)
```

```python
user = 'DCbatman'
limit=2900

tweets = tweepy.Cursor(api.user_timeline, screen_name=user, count=200, tweet_mode='extended').items(limit)

# tweets = api.user_timeline(screen_name=user, count=limit, tweet_mode='extended')

# create DataFrame
columns = ['User', 'Tweet', 'time']
data = []

for tweet in tweets:
    data.append([tweet.user.screen_name, tweet.full_text, tweet.created_at])

df = pd.DataFrame(data, columns=columns)

print(df)
```

```python
df
```

```python
# saved as df.to_csv('DCbatman_timeline.csv')
```

```python
keywords = '@DCbatman'  # for user

limit=2900

tweets = tweepy.Cursor(api.search_tweets, q=keywords, count=100, tweet_mode='extended').items(limit)

columns = ['User', 'Tweet', 'Time']
data = []

for tweet in tweets:
    data.append([tweet.user.screen_name, tweet.full_text, tweet.created_at])

batman_df = pd.DataFrame(data, columns=columns)
```

```python
batman_df
```

```python
batman_df.to_csv('batman_data.csv')
```
