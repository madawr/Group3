```python
import pandas as pd

import datetime

import numpy as np

import matplotlib.pyplot as plt

from pandas.plotting import scatter_matrix

#!pip install yfinance

import yfinance as yf

%matplotlib inline
```

```python
start = "2022-02-14"
end = '2022-03-18'
cnk = yf.download('CNK',start,end)
amc = yf.download('AMC',start,end)
cinel = yf.download('CINE.L',start,end)
```

```python
cnk['Open'].plot(label = 'CNK', figsize = (15,7))
amc['Open'].plot(label = 'AMC')
cinel['Open'].plot(label = 'CINE.L')
plt.title('Stock Prices of CNK, AMC and CINE.L')
plt.legend()
```

```python
cnk['Volume'].plot(label = 'CNK', figsize = (15,7))
amc['Volume'].plot(label = 'AMC')
cinel['Volume'].plot(label = 'CINE.L')
plt.title('Volume of Stock traded')
plt.legend()
```

```python
#Market Capitalisation
cnk['MarktCap'] = cnk['Open'] * cnk['Volume']
amc['MarktCap'] = amc['Open'] * amc['Volume']
cinel['MarktCap'] = cinel['Open'] * cinel['Volume']
cnk['MarktCap'].plot(label = 'CNK', figsize = (15,7))
amc['MarktCap'].plot(label = 'AMC')
cinel['MarktCap'].plot(label = 'CINE.L')
plt.title('Market Cap')
plt.legend()
```

```python
#Volatility
cnk['returns'] = (cnk['Close']/cnk['Close'].shift(1)) -1
amc['returns'] = (amc['Close']/amc['Close'].shift(1))-1
cinel['returns'] = (cinel['Close']/cinel['Close'].shift(1)) - 1
cnk['returns'].hist(bins = 100, label = 'CNK', alpha = 0.5, figsize = (15,7))
amc['returns'].hist(bins = 100, label = 'AMC', alpha = 0.5)
cinel['returns'].hist(bins = 100, label = 'CINE.L', alpha = 0.5)
plt.legend()
```
